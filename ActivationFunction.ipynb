Sigmoid fuction gives the value between 0 & 1

```python
import math

def singmoid(x):
    return 1 / (1 + math.exp(-x))

print(singmoid(3))
```

Tan h gives the values between -1 & 1

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

print(tanh(3))
```

Relu gives the max val between 0 & x

```python
def relu(x):
  return max(0, x)

print(relu(-3))
print(relu(33))
```

Leaky relu gives the value directly if the value greater the 0 and if not the it does 0.1 * x

```python
def leakyrelu(x):
  return max(0.1*x, x)

print(leakyrelu(-3))
print(leakyrelu(33))
```
